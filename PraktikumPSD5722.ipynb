```python
#import modules

import pandas as pd
import numpy as np
```

```python
#persiapan data frame
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,10, size=(n_rows, n_cols)),columns=cols)

df
```

```python
df.add_prefix('kolom')
```

```python
df.add_suffix('kolom')
```

```python
#seleksi column dengan logika or

df[(df['A'] == 9)]
```

```python
#seleksi column dengan fungsi isin

df[df['A'].isin([9])]
```

```python
#memanfaatkan fungsi not dalam isin dengan operator negasi

df[~df['A'].isin([9])]
```
