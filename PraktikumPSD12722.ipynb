```python
#import modules

import pandas as pd
import numpy as np
```

```python
#konversi tipe data string ke numerik

data = {'col 1': ['1','2','3', 'teks'],
        'col 2': ['1','2','3','4']}

df_lain = pd.DataFrame(data)
df_lain
```

```python
df_lain.dtypes
```

```python
#konversi tipe data dengan fungsi astype

df_x = df_lain.astype({'col 2': 'int'})
df_x
```

```python
df_x.dtypes
```

```python
#konversi tipe data ke tipe data numerik

df_lain.apply(pd.to_numeric, errors='coerce')

#coerce, kalo ada data yg gabisa dikonversi ke numerik, data tsb bakal jadi NaN
```
