```python
import pandas as pd
import numpy as np
```

```python
n_rows = 10
n_cols = 5
cols = tuple ('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)

df
```

```python
data = {'day' : [1,2,10,25,12],
        'month' : [1,2,4,5,6],
        'year' : [2001,2010,2015,2019,2020]}

df = pd.DataFrame(data)
df
```

```python
n_rows = 10
n_cols = 1
cols = ('usia',)

df_usia = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)),
                       columns=cols)
df_usia
```

```python
#Pengelompokan nilai numerik ke dalam beberapa kategori menggunakan cut

df_usia['kelompok_usia'] = pd.cut(df_usia['usia'],
                                bins = [0, 18, 65, 99],
                                labels = ['anak', 'dewasa', 'manula'])
df_usia
```

```python
n_rows = 5
n_cols = 5
cols = tuple ('ABCDE')

df_a = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df_a
```

```python
# Pecah data frame menjadi 2

df1 = df_a.copy(deep=True)
df1 = df1.drop([1,4])
df1
```

```python
df2 = df_a.copy(deep=True)
df2 = df_a.drop([0,3])
df2
```

```python
# Menggunakan metode merge untuk menggabungkan 2 df
# Inner di kedua dataframe datanya ada semua
# index reset kembali menjadi 0

df_inner = pd.merge(df1,df2,how='inner')
df_inner
```

```python
df_outer = pd.merge(df1,df2,how='outer')
df_outer
```
