```python
import pandas as pd
import numpy as np
```

```python
n_rows = 5
n_cols = 2
cols = ['omset', 'operasional']

df = pd.DataFrame(np.random.randint(1,20, size=(n_rows, n_cols)), columns=cols)
df
```

```python
# Underscore dipakai untuk ignore character
# Omset konversi/kalikan ke dalam satuan ratusan ribu, operasional per sepuluh ribu
df['omset'] = df['omset'] * 100_000
df['operasional'] = df['operasional'] = 100_000
df
```

```python
#membentuk kolom tanggal, generate index menggunakan date time
#frequencynya daily/harian

df.index = pd.util.testing.makeDateIndex(n_rows, freq='D')
df = df.reset_index()
df = df.rename(columns={'index':'tanggal'})

df
```

**Melakukan Formatting Tampilan Data Frame**

```python
#sesuaikan dengan variabel data frame, urutan tidak berpengaruh, key harus sesuai nama kolom
#formatting disampaikan dalam tipe data string, maka diapit oleh tanda kutip
#.2f menyertakan dua digit extra di belakang koma

formatku = {'tanggal':'{:%d/%m/%y}',
            'operasional':'Rp {:.2f}',
            'omset':'Rp {:.2f}'}
laporan = df.style.format(formatku)
laporan
```

```python
#tipe data (objectnya styler)
#styler tidak bisa melakukan hal-hal yang bisa dilakukan dataframe

type(laporan)
```

```python
#formatting tanpa objek
laporan.hide_index()
```

```python
#menyertakan judul dari frame
laporan.set_caption('Data Omset dan Operasional')
```

```python
laporan.highlight_min('omset', color='pink')
laporan.highlight_max('omset', color='lightgreen')

laporan.highlight_min('operasional', color='lightblue')
laporan.highlight_max('operasional', color='grey')
```

**Menggabungkan (merge) dua Data Frame secara berdampingan**

```python
d1 = {'col1':[1, 2, 3],
      'col2':[10, 20, 30]}
df1 = pd.DataFrame(d1)
df1
```

```python
d2 = {'col1':[4, 5, 6],
      'col2':[40, 50, 60]}
df2 = pd.DataFrame(d2)
df2
```

```python
#merging yang dilakukan sifatnya ke samping, ada 4 kolom
#parameter left & right

df3 = pd.merge(df1, df2, left_index=True, right_index=True)
df3
```

**Agregasi pada Pandas Data Frame dengan fungsi agg()**

```python
data_baru = pd.read_csv('./Iris.csv')
data_baru
```

```python
#Mengenal fungsi groupby() dan fungsi agregasi

data_baru.groupby('Species')['PetalLengthCm'].count().to_frame()

#Dikelompokkan berdasarkan species, di kelompok petalLength dikenakan fungsi agregasi
#count = menghitung jumlah baris
#to_frame() = membuat baris baru
```

```python
data_baru.groupby('Species')['PetalLengthCm'].mean().to_frame
#setelah dikelompokkan dihitung rata-rata
```

```python
data_baru.groupby('Species')['PetalLengthCm'].median().to_frame
#setelah dikelompokkan dihitung median
```

```python
#menggunakan agg untuk multiple function aggregation
data_baru.groupby('Species')['PetalLengthCm'].agg(['count', 'median', 'mean'])
```

```python
#agregasi dengan describe()
data_baru.groupby('Species')['PetalLengthCm'].describe()
```
