**Memecah nilai string dari suatu kolom ke dalam beberapa kolom**

```python
import pandas as pd
```

```python
# persiapan data frame dictionary

data = {'nama': ['Farida Putri', 'Sariza Nur', 'Khairina Rahma'],
        'tempat_kelahiran': ['Bogor, Jawa Barat', 'Semarang, Jawa Tengah', 'Magetan, Jawa Timur']}

df = pd.DataFrame(data)
df
```

```python
# mau memecah kolom menjadi nama depan dan nama belakang
# displit berdasarkan spasi
# pararel expand=True untuk mengizinkan output berbentuk kolom/multiply column

df[['nama_depan', 'nama_belakang']] = df['nama'].str.split(' ', expand=True)
df
```

```python
#memecah nama kota dan provinsi
#yang membedakan dengan yang di atas adalah dengan memakai kota

df[['kota', 'provinsi']] = df['tempat_kelahiran'].str.split(',', expand=True)
df
```

**Menggunakan unstack() pada Multiple Indexes Data Frame**

```python
#menata ulang data frame dengan multiple indexes menggunakan unstack()

import pandas as pd
```

```python
df = pd.read_csv('./titanicfull.csv')
df.head()
```

```python
# mengelompokkan data ke dalam beberapa kelompok, di sini dikelompokkan berdasarkan sex dan pclass
# data yang akan diamati yaitu data yang survivednya 1 untuk selamat dan 0 untuk tidak selamat
# rata-rata nilai yang selamat
# to_frame untuk mempercantik tampilan

df.groupby(['sex', 'pclass'])['survived'].mean().to_frame()
```

```python
# menata ulang data frame dengan multiple indexes
# cara penempatan index sudah berbeda

df.groupby(['sex', 'pclass'])['survived'].mean().unstack()
```

**Resampling pada data deret waktu(time series)**

```python
import pandas as pd
import numpy as np
```

```python
#365 untuk hari, 24 untuk jam
# interval waktu berdasarkan jam

n_rows = 365 * 24
n_cols = 2
cols = ['kolom1', 'kolom2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq = 'H')

df
```

```python
# resampling data dengan interval monthly dengan .resample
# interval waktu berdasarkan bulan

df.resample('M')['kolom1'].sum().to_frame()
```

```python
# resampling data dengan interval daily
# interval waktu berdasarkan hari

df.resample('D')['kolom1'].sum().to_frame()
```
